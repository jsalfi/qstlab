```python
import matplotlib.pyplot as plt
from transmission import *
```

# Bragg Reflectors

```python
# center frequencies of the inductors
f1 = 5e9
f2 = 5e10

# wavelengths
wl1 = c/f1
wl2 = c/f1

# physical lengths
l1 = wl1/4
l2 = wl2/4

#matched impedance assumed to be satisfied

# number of inductors in bragg reflector
N = 7

# generate wavenumbers
n_pts = 1000
freqs = np.linspace(1e9, 10e9)
Bs = 2*np.pi*freqs/c

# set characteristic impedances of inductors
Z1 = 100
Z2 = Z1*4.25

# abcd matrices of the inductors
low = abcd_series(Z1, Bs, l1)
high = abcd_series(Z2, Bs, l2)

# cascade the network
a = cascade_series(N, low, high, low_first=True)

# plot the S parameters
plt.plot(freqs/1e9, 20*np.log10(np.abs(a.S21())), label="$S_{21}$")
plt.plot(freqs/1e9, 20*np.log10(np.abs(a.S11())), label="$S_{11}$")
plt.ylim([-50, 5])
plt.legend()
plt.xlabel("f [GHz]")
plt.ylabel("[dB]")
```

# Shunts

```python
# center frequencies of the inductor
f1 = 5e9

# wavelength
wl1 = c/f1

# physical length
l1 = wl1/4

#matched impedance assumed to be satisfied

# number of inductors in stub
N = 3

# generate wavenumbers
n_pts = 1000
freqs = np.linspace(1e9, 10e9, n_pts)
Bs = 2*np.pi*freqs/c

# set characteristic impedances of inductor
Z1 = 100

# abcd matrices of the inductor
resonator = abcd_series(Z1, Bs, l1)
shunt = abcd_shunt(Z1, Bs, l1)

# cascade the network
a = cascade_stub(N, shunt, resonator)

# plot the S parameters
plt.plot(freqs/1e9, 20*np.log10(np.abs(a.S21())), label="$S_{21}$")
plt.plot(freqs/1e9, 20*np.log10(np.abs(a.S11())), label="$S_{11}$")
plt.ylim([-100, 5])
plt.legend()
plt.xlabel("f [GHz]")
plt.ylabel("[dB]")
```
